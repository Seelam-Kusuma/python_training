```python
faults=(fed['player1 double faults'])+(fed['player2 double faults'])
total_win=fed['player1 total points won']+fed['player2 total points won']
proportion=faults/total_win
proportion
 
```

```python
average_wins = fed.groupby('surface')['player2 total points total'].mean()
average_wins
```
